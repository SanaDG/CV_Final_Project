# Chess Position Detection

# Objective
The objective of this project is to build a machine learning model that can detect the positions of chess pieces on a board from images. This involves training a Convolutional Neural Network (CNN) to classify the chess pieces based on their positions in the given images.

Here we trained and evaluate two models.

# Dataset
The dataset consists of images of chess positions split into training and testing sets. Each image is labeled with the positions of chess pieces on the board.
https://www.kaggle.com/datasets/koryakinp/chess-positions

# Content
100000 images of a randomly generated chess positions of 5-15 pieces (2 kings and 3-13 pawns/pieces)
Images were generated using 28 styles of chess boards and 32 styles of chess pieces totaling 896 board/piece style combinations.

Images were generated using this custom-build tool

All images are 400 by 400 pixels.

Training set: 80000 images
Test set: 20000 images
Pieces were generated with the following probability distribution:

30% for Pawn
20% for Bishop
20% for Knight
20% for Rook
10% for Queen
2 Kings are guaranteed to be on the board.

Labels are in a filename in Forsyth–Edwards Notation format, but with dashes instead of slashes.

### Feature engineering

```python
def image_to_squares(img,heights,widths):
  squares = []
  for i in range(0,8):
    for j in range(0,8):
      squares.append(img[i*heights//8:i*heights//8+heights//8,j*widths//8:j*widths//8+widths//8])
  return np.array(squares)
```

```python
def preprocess_image(img_path):
    height =240
    width =240

    # change to Grey scal
    img = cv2.imread(img_path, cv2.COLOR_BGR2GRAY)
  
    # resize the image to the desired size
    gray_image = cv2.resize(img, (width, height))
    
    # Normalize the image
    gray_image =(gray_image - np.min(gray_image)) / (np.max(gray_image) - np.min(gray_image))

    squares = image_to_squares(gray_image,height,width)
    return squares
```

- Create a FEN Label Encode /Decoder Functions
- **NOTE** These 2 Functions are taken from this Kaggle notebook : https://www.kaggle.com/code/koryakinp/chess-fen-generator/notebook

```python
def onehot_from_fen(fen):
    eye = np.eye(13)
    output = np.empty((0, 13))
    fen = re.sub('[-]', '', fen)

    for char in fen:
        if(char in '12345678'):
            output = np.append(
              output, np.tile(eye[12], (int(char), 1)), axis=0)
        else:
            idx = piece_symbols.index(char)
            output = np.append(output, eye[idx].reshape((1, 13)), axis=0)

    return output

def fen_from_onehot(one_hot):
    output = ''
    for j in range(8):
        for i in range(8):
            if one_hot[j][i] == 12:
                output += ' '
            else:
                output += piece_symbols[one_hot[j][i]]
        if(j != 7):
            output += '-'

    for i in range(8, 0, -1):
        output = output.replace(' ' * i, str(i))

    return output
```

```python
def train_gen(features):
    for i, img in enumerate(features):
        y = onehot_from_fen(get_image_FEN_label(img))
        x = preprocess_image(img)
        yield x, y

def pred_gen(features):
    for i, img in enumerate(features):
        y = onehot_from_fen(get_image_FEN_label(img))
        x = preprocess_image(img)
        yield x, y
```

### Data Loading and Preprocessing

```python
# Directory paths
train_dir = "/kaggle/input/chess-positions/train"
test_dir = "/kaggle/input/chess-positions/test"

# List the files in the directories
train_files = os.listdir(train_dir)
test_files = os.listdir(test_dir)

# Check file extensions (make sure to include hidden files check)
train_jpeg_files = [f for f in train_files if f.lower().endswith('.jpeg')]
test_jpeg_files = [f for f in test_files if f.lower().endswith('.jpeg')]

# Get the paths of images using absolute paths
train_path = [os.path.join(train_dir, f) for f in train_jpeg_files]
test_path = [os.path.join(test_dir, f) for f in test_jpeg_files]

# Shuffle train and test samples
np.random.shuffle(train_path)
np.random.shuffle(test_path)

# Decide the number of train and test samples
train_size = 4000
test_size = 800
train = train_path[:train_size]
test = test_path[:test_size]
```

```python
# Function to preprocess images for ResNet50
def preprocess_image_resnet(img_path):
    img = cv2.imread(img_path)  # Read image
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    img = cv2.resize(img, (224, 224))  # Resize to 224x224 (ResNet50 input size)
    img = preprocess_input(img)  # Preprocess input based on ImageNet mean normalization
    return img


train_images_resnet = np.array([preprocess_image_resnet(img_path) for img_path in train])
test_images_resnet = np.array([preprocess_image_resnet(img_path) for img_path in test])

piece_symbols = 'prbnkqPRBNKQ'
```

### Resnet50 model Training

```python
def extract_piece_info(file_path):
    file_name = os.path.basename(file_path)
    color = "White" if "white" in file_name.lower() else "Black"
    piece_type = next((piece for piece in piece_symbols if piece in file_name), None)
    return color, piece_type

# Function to get FEN label from file path
def get_image_FEN_label(image_path):
    fen_label = image_path.replace('.jpeg', '').split('/')[-1]
    return fen_label

# Function to convert FEN notation to one-hot encoding
def onehot_from_fen(fen):
    eye = np.eye(13)
    output = np.empty((0, 13))
    fen = fen.replace('-', '')

    for char in fen:
        if char.isdigit():
            output = np.append(output, np.tile(eye[12], (int(char), 1)), axis=0)
        else:
            idx = piece_symbols.index(char)
            output = np.append(output, eye[idx].reshape((1, 13)), axis=0)

    return output

# Get labels for training and testing images
train_labels = np.array([onehot_from_fen(get_image_FEN_label(img_path)) for img_path in train])
test_labels = np.array([onehot_from_fen(get_image_FEN_label(img_path)) for img_path in test])

# Build ResNet50 model
def build_resnet50_model(num_classes):
    # Load ResNet50 base model without the top (fully connected) layers
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    
    # Add your custom layers on top of ResNet50
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation='relu')(x)
    
    # Adjust the output layer to produce predictions for each square on the chessboard
    predictions = Dense(64 * num_classes, activation='softmax')(x)
    predictions = Reshape((64, num_classes))(predictions)
    
    # Create model
    model = Model(inputs=base_model.input, outputs=predictions)
    
    # Freeze layers in the base model
    for layer in base_model.layers:
        layer.trainable = False
    
    # Compile the model
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    
    return model
```

```python
# Assuming you have defined train, test, train_labels, and test_labels appropriately
# Build the ResNet50 model
model = build_resnet50_model(num_classes=13)

# Train the model
history = model.fit(
    train_images_resnet, train_labels,
    epochs=100,
    validation_data=(test_images_resnet, test_labels),
    batch_size=32
)

# Evaluate the model
loss, accuracy = model.evaluate(test_images_resnet, test_labels)
print(f"Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}")
```
